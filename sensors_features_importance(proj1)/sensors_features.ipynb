# Some techniques of best feature selection for our ML model:

(1)   a.  Forward selection: add features in every iteration and evaluate the accuracy of the model.
        
        b. Backward selection: train the model with all the features and start reducing the number of features  in each iteration 
        
        c. Recursive selection: this is a permutation of features and the iteration becomes the factorinal problem. for n features, (n)! of iterations are needed wchich not practical for big datasets.
        
(2)  a. univeriated selection, b. Feature importance and  c. Embedded methods, Filters Methods, and Wrappers.
        
Usually, techique 2 is the preffered one as it reduces the computation resources and can be implemented using statistical tools of scikit.
            

```python
# Lets get the features from a data frame
train_features = sensors_df.iloc[:,2:12]
train_features
print("training_features:")
train_features
```

```python
# read class labels from the data frame
y_labels = sensors_df.iloc[:,1]
print('training class labels')
y_labels
```

# Analyzing the features(sensors data) using statistical tools

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# get the statistical score of the features and then sort them features using their score.
best_features = SelectKBest(score_func=chi2, k=5).fit(train_features, y_labels)
```

```python
# get the scores of the features(sensors)
scores = best_features.scores_
# get the name of the columns or sensor names
column_names =train_features.columns

# and put column_names in data frame
df = pd.DataFrame(column_names,columns={'sensor_name'})
# add the score into data frame
df['score']= scores
print("The unsorted score of features or sensors:\n")
df
```

Notes: from the above table has unsorted score. so, lets sort them. After sorting, the most predictive features comes from the following sensors sorted in descending(best to worst) order shown the code(table) below.

```python
sorted_features = df.sort_values(by=['score'],ascending=False)
print("The sorted score of features or sensors:\n")
sorted_features
```

# consclusion 1: "From the above table,sensor 8 has the most predictive power(feature). One the other hand, sensor 6 is the least predictive feature."

# The Machine Learing(ML) Part: create a model, train, and test the model using the features given.

```python
#observe the shape of  the  data
print("x shape:",train_features.shape)

print("y shape:",y_labels.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from warnings import simplefilter     
    

# disable warnings 
def ignore_warning_func():
    # ignore all future warnings
    simplefilter(action='ignore', category=FutureWarning)
    

ignore_warning_func()

#split out data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(train_features, y_labels, test_size=0.3, random_state=42)
model = LogisticRegression().fit(x_train, y_train)

y_predictions = model.predict(x_test)
print('predictions\n',y_predictions)

# create a data frame with sensor names
new_df= pd.DataFrame(column_names,columns={'sensor_name'})
    
# add the scores or feature importance to the data frame 
new_df['score']= model.coef_[0]

# sort the feature importance (score) of the sensors
sorted_df = new_df.sort_values(by='score',ascending=False)
sorted_df
```

# conclusion 2: Using Logistic Regression model, the model predicted the feature importance that  'sensor8' has  the most predictive power(feature) while the other sensors are relatively less powerful that 'sensor8'.

```python
# Evaluating  the model's performance using testing data sets
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score
import numpy as np

y_pred =np.rint(y_predictions)

cm =  confusion_matrix(y_test,y_pred)

print("The confusion matrix:\nNB:the diagonal entries of the matrix are the correct prediction.\n",cm) # the diagonal elements are the correct predictions
print("the number of test samples:{}".format(y_test.shape[0]))

acc = accuracy_score(y_test, y_pred)    # add ', normalize = False' to see the actual number of predictions.
print("the accuracy of the model: {:.2f}%".format(acc*100))

print("Precision:{:.2f}%".format(precision_score(y_test, y_pred)*100))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:{:.2f}%".format(recall_score(y_test, y_pred)*100))
```

# The model's performance seems pretty reasonable though the precision score is a little bit low. Over all, it seems okay.

```python
import  matplotlib.pyplot as plt
import numpy as np
features_importance = model.coef_[0]
scores=[]
sensors=[]

for k , score in enumerate(features_importance) :
    scores.append(np.round(score,3))
    sensors.append(k)
# plot feature importance of the logistic regression model in bar chart

plt.bar(sensors,scores)
plt.ylabel('features score')
plt.xlabel('sensor')
plt.title('feature importance from a logistic regression  model')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(100,criterion='entropy') 
  
# Training the model 
rf_model.fit(x_train,y_train) 
  
# Computing the importance of each feature 
feature_importance = rf_model.feature_importances_ 
randomforest_pred= rf_model.predict(x_test)
acc = accuracy_score(y_test, randomforest_pred)    # add ', normalize = False' to see the actual number of predictions.
print("\n\nthe accuracy of the model: {:.2f}%".format(acc*100))

print("Precision:{:.2f}%".format(precision_score(y_test, randomforest_pred)*100))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:{:.2f}%".format(recall_score(y_test,randomforest_pred)*100))

# lets sort the features importance obtained from random forest classifier

randomforest_df = pd.DataFrame(column_names,columns={'sensor_name'})
randomforest_df['score']= feature_importance
# sorted features importance
sorted_rf_df = randomforest_df.sort_values(by='score', ascending=False)
sorted_rf_df
```

```python
# plot feature importance of the random forest classifier  model in bar chart

rf_features_importance = rf_model.feature_importances_
rf_scores=[]
rf_sensors=[]

for i , score in enumerate(rf_features_importance) :
    rf_scores.append(np.round(score,3))
    rf_sensors.append(i)
plt.bar(rf_sensors,rf_scores)
plt.ylabel('features score')
plt.xlabel('sensor')
plt.title('feature importance from random forest classifier model')
plt.show()
```
